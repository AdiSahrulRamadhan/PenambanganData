# Memahami data

## Apa itu Iris-datasets ?

Dataset Iris adalah kumpulan data yang sangat umum digunakan untuk pembelajaran mesin dan statistik. Dataset ini diperkenalkan oleh ahli biologi yang bernama Sir Ronald Fisher pada tahun 1936 yang saat itu juga pertama kali diperkenalkannya dan dalam analisis diskriminan linier. Pada iris dataset ini terdiri dari pengukuran panjang dan lebar kelopak dan mahkota dari tiga spesies bunga iris: Iris setosa, Iris versicolor, dan Iris virginica. Setiap bunga diukur pada empat variabel: panjang kelopak, lebar kelopak, panjang mahkota, dan lebar mahkota. Iris dataset digunakan secara luas sebagai contoh untuk mengajarkan berbagai algoritma pembelajaran mesin, terutama untuk tugas klasifikasi. Tujuan umumnya adalah untuk memprediksi spesies bunga iris berdasarkan pengukuran fitur-fiturnya. Karena dataset ini relatif kecil dan mudah dipahami, serta memiliki struktur yang terdefinisi dengan baik sehingga mudah dipahami.

Data Understanding leadalah tahap awal dalam proses analisis data di mana analis mencoba memahami dataset yang akan diolah. Tujuan utama dari tahap ini adalah untuk mendapatkan wawasan awal tentang struktur, kualitas, dan karakteristik data yang akan dijalankan. dibawah ini langkah-langkahnya.

2. Periksa jumlah data dan data yang hilang.
3. Hitung statistik deskriptif untuk memahami distribusi data.
4. Buat visualisasi data untuk menemukan pola dan tren.
5. Analisis korelasi antar variabel.
6. Validasi keabsahan data.
7. Dokumentasikan temuan untuk referensi selanjutnya.

## Penjelasan Fitur-fitur dalam Data

### Pada Iris Dataset memiliki 4 fitur atau variabel dibawah ini penjelasannya.

1. Panjang Kelopak (sepal length): Ini adalah panjang dari kelopak bunga iris, diukur dari pangkal hingga ujung.
2. Lebar Kelopak (sepal width): Ini adalah lebar dari kelopak bunga iris, diukur pada bagian terlebar kelopak.
3. Panjang Mahkota (petal length): Ini adalah panjang dari mahkota bunga iris, diukur dari pangkal hingga ujung.
4. Lebar Mahkota (petal width): Ini adalah lebar dari mahkota bunga iris, diukur pada bagian terlebar mahkota.

Dalam dataset iris, terdapat beberapa kolom atau atribut yang menggambarkan data bunga iris. Biasanya, terdapat lima kolom dalam dataset iris, termasuk empat kolom yang menyatakan fitur-fitur yang telah disebutkan sebelumnya, dan satu kolom yang menyatakan spesies bunga iris. Berikut adalah penjelasan untuk setiap kolom:

- Id : ini digunakan untuk imembantu mengidentifikasi dan merujuk kembali record / entri tertentu dalam dataset tipe datanya integer.

- Sepal Length (panjang kelopak): Kolom ini berisi nilai-nilai yang mewakili panjang dari kelopak bunga iris dalam satuan yang relevan (misalnya, centimeter).

- Sepal Width (lebar kelopak): Kolom ini berisi nilai-nilai yang mewakili lebar dari kelopak bunga iris dalam satuan yang relevan.

- Petal Length (panjang mahkota): Kolom ini berisi nilai-nilai yang mewakili panjang dari mahkota bunga iris dalam satuan yang relevan.

- Petal Width (lebar mahkota): Kolom ini berisi nilai-nilai yang mewakili lebar dari mahkota bunga iris dalam satuan yang relevan.

- Species (spesies): Kolom ini berisi label atau kategori yang menunjukkan spesies dari bunga iris yang sesuai dengan pengukuran yang terdapat dalam empat kolom sebelumnya. Biasanya, terdapat tiga nilai kategori: "Iris-setosa", "Iris-versicolor", dan "Iris-virginica".

### Metode Knn

K-Nearest Neighbors (KNN) adalah  algoritma pembelajaran mesin yang digunakan  untuk klasifikasi dan regresi.Ide dasar dari algoritma KNN adalah mengklasifikasikan atau memprediksi  titik data dengan mempertimbangkan mayoritas label kelas dari k titik data terdekat (nearest neighbours) pada ruang fitur. Jarak antara titik data prediksi dan titik data tetangga  dihitung, dan titik data terdekat digunakan sebagai dasar prediksi.

Langkah-langkah Metode Knn

- Menentukan Parameter K: Pilih jumlah tetangga terdekat (K) yang akan digunakan untuk membuat prediksi.

- Menghitung Jarak: Gunakan metrik jarak (seperti Euclidean, Manhattan, atau Minkowski) untuk mengukur jarak antara titik data yang akan diprediksi dengan setiap titik data dalam set pelatihan.

- Menemukan Tetangga Terdekat: Identifikasi K titik data terdekat (tetangga terdekat) dari titik data yang akan diprediksi berdasarkan jarak yang telah dihitung.

- Menghitung Mayoritas Kelas: Tentukan mayoritas kelas dari tetangga terdekat. Jika K adalah ganjil, mayoritas kelas dapat ditentukan langsung; jika K adalah genap, teknik bobot jarak dapat digunakan.

- Membuat Prediksi: Tetapkan kelas mayoritas sebagai prediksi untuk titik data yang akan diprediksi.

```
import mysql.connector
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

conn = mysql.connector.connect(
    host="database2-adisahrulramadhan22072.a.aivencloud.com",
    user="avnadmin",
    password="AVNS_o9Uf-iAsqNySoqOtmNf",
    port = "23413",
    database="db_mysql"
)

# mengambil semua data sepalLength yang bukan missing value
query = "SELECT *FROM perhitunganKnn WHERE SepalLengthCm IS NOT NULL"
dataset = pd.read_sql(query, conn)

# Menutup koneksi ke database
conn.close()

# Memisahkan fitur dan label
feature_columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
X = dataset[feature_columns].values
y = dataset['Species'].values

# Mengubah label menjadi numerik
le = LabelEncoder()
y = le.fit_transform(y)

# Membagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Membuat model KNN dengan k=5
classifier = KNeighborsClassifier(n_neighbors=5)

# Melatih model
classifier.fit(X_train, y_train)

# Memprediksi data uji
y_pred = classifier.predict(X_test)

# Menghitung akurasi
accuracy = accuracy_score(y_test, y_pred) * 100
print('Accuracy of our model is equal ' + str(round(accuracy, 2)) + ' %.')

# Membuat list k untuk KNN
k_list = list(range(1, 50, 2))
cv_scores = []

# Melakukan 10-fold cross validation
for k in k_list:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=10, scoring='accuracy')
    cv_scores.append(scores.mean())

# Menghitung MSE
MSE = [1 - x for x in cv_scores]

plt.figure(figsize=(15, 10))
plt.title('The optimal number of neighbors', fontsize=20, fontweight='bold')
plt.xlabel('Number of Neighbors K', fontsize=15)
plt.ylabel('Misclassification Error', fontsize=15)
sns.set_style("whitegrid")
plt.plot(k_list, MSE)
plt.show()

# Menemukan nilai k terbaik
best_k = k_list[MSE.index(min(MSE))]
print("The optimal number of neighbors is %d." % best_k)
```

## Deteksi Outlier (Lokal Outlier Factor)

### Konsep Local Outlier Faktor

Local Outlier Factor (LOF) adalah  metode  deteksi anomali yang digunakan untuk mengidentifikasi titik data yang  outlier dalam konteks lokalnya. LOF mengukur seberapa tidak biasa suatu titik data muncul dibandingkan dengan titik data tetangganya dalam ruang fitur. Metode ini memungkinkan untuk menemukan anomali terkait kepadatan data lokal. Artinya suatu titik data mungkin merupakan outlier di satu area, namun normal di area lain.

### Menhitung manual Local Outlier Factor

Data contoh data dengan tipe data numeric 3 fitur jumlah data 10. K = 3 tau jarak tetangga terdekat dari 10 objek. P = 2 Metode (Eclaudien Distance).

Berikut langkah-langkah mendeteksi outlier dengan metode LOF :

- Menghitung jarak dan menentukan tetangganya.

- Menghitung Reachability Distance (RD)

- Menghitung Local Reachability Distance (LRD)

- Menghitung Local Outlier Factor (LOF)

Sebelum menentukan data yang akan kita gunakan untuk melakukan perhitungan manual. Contoh data memiliki 3 fitur yaitu X, Y, Z yang masing masing bertipe data numeric, dan berjumlah 10. Dibawah ini penjelasan konsep setiap langkah-langkah perhitungan yang harus kita pahami dalam memahami LOF. Sumber referensi materi yang saya pelajari dari situs link ini supaya lebih paham lagi https://towardsdatascience.com/local-outlier-factor-lof-algorithm-for-outlier-identification-8efb887d9843

- K-distance and K-neighbors

- Reachability distance (RD)

- Local reachability density (LRD)

- Local Outlier Factor (LOF)

1. K-distance and K-neighbors

K-jarak adalah jarak antara suatu titik, dan merupakan tetangga terdekat Kᵗʰ. Tetangga K yang dilambangkan dengan Nₖ(A) mencakup himpunan titik yang terletak di dalam atau pada lingkaran berjari-jari jarak K. K-tetangga bisa lebih dari atau sama dengan nilai K. Bagaimana mungkin?

Kita akan melihat contohnya. Katakanlah kita mempunyai empat titik A, B, C, dan D (ditunjukkan di bawah).

<img src="image-20240320-202655.png" width="actual" align="" />

Jika K=2, K-tetangga A adalah C, B, dan D. Di sini, nilai K=2 tetapi ||N₂(A)|| = 3. Oleh karena itu, ||Nₖ(titik)|| akan selalu lebih besar atau sama dengan K.

2. Reachability distance (RD)

Dibawah ini didefinisikan sebagai jarak K maksimum Xj dan jarak antara Xi dan Xj. Ukuran jarak bersifat khusus untuk masalah (Euclidean, Manhattan, dll.)

<img src="image-20240320-203005.png" width="" align="" />

<img src="image-20240320-203032.png" width="" align="" />

Dalam istilah awam, jika titik Xi terletak di dalam K-tetangga Xj, maka jarak jangkauannya adalah K-jarak Xj (garis biru), jika tidak, jarak jangkauannya adalah jarak antara Xi dan Xj (garis oranye).

3. Local reachability density (LRD)

<img src="image-20240320-203128.png" width="" align="" />

LRD merupakan kebalikan dari rata-rata jarak jangkauan A dari tetangganya. Secara intuitif berdasarkan rumus LRD, semakin besar jarak jangkauan rata-rata (yaitu, tetangga jauh dari titik tersebut), maka semakin sedikit kepadatan titik yang ada di sekitar titik tertentu. Ini menunjukkan seberapa jauh suatu titik dari kelompok titik terdekat.Nilai LRD yang rendah menunjukkan bahwa cluster terdekat berada jauh dari titik.

4. Local Outlier Factor (LOF)

<img src="image-20240320-203452.png" width="" align="" />

LRD tiap titik digunakan untuk membandingkan dengan rata-rata LRD K tetangganya. LOF adalah perbandingan rata-rata LRD K tetangga A terhadap LRD A.

Secara intuitif, jika suatu titik bukan merupakan pencilan (inlier), rasio rata-rata LRD tetangganya kira-kira sama dengan LRD suatu titik (karena kepadatan suatu titik dan tetangganya kira-kira sama). Dalam hal ini, LOF hampir sama dengan 1. Sebaliknya, jika suatu titik merupakan outlier, LRD suatu titik lebih kecil dari rata-rata LRD tetangganya. Maka nilai LOF akan tinggi.

Umumnya jika LOF > 1 maka dianggap outlier, namun hal tersebut tidak selalu benar. Katakanlah kita mengetahui bahwa kita hanya memiliki satu outlier dalam data, lalu kita ambil nilai LOF maksimum di antara semua nilai LOF, dan titik yang sesuai dengan nilai LOF maksimum akan dianggap sebagai outlier.

### Menghitung menggunakan Python (Scikit-learn)

Setelah menentukan outlier manual menggunakan excel, berikutnya kita akan menguji hasil tersebut menggunakan scikit-learn.

```
from sklearn.neighbors import LocalOutlierFactor
import numpy as np

X = [
    (1, 4, 1),
    (0, 2, 1),
    (2, 1, 1),
    (1, 1, 2),
    (2, 3, 2),
    (25, 25, 25),
    (4, 0, 0),
    (1, 1, 2),
    (3, 3, 1),
    (2, 1, 3),
]

lof = LocalOutlierFactor(n_neighbors=3, contamination=0.1)

y_pred = lof.fit_predict(X)

print("Predicted table:", y_pred)
print("Negative LOF scores:", -lof.negative_outlier_factor_)  # Digunakan untuk ubah tanda negatif untuk mendapatkan nilai LOF positif
```

Ini memberikan pemahaman yang lebih dalam tentang karakteristik outlier dalam dataset, di mana mereka mungkin menonjol dalam satu dimensi tetapi tidak dalam dimensi lainnya.

```
!pip install mysql-connector-python
from sklearn.neighbors import LocalOutlierFactor
import mysql.connector
import matplotlib.pyplot as plt

hostname = "database2-adisahrulramadhan22072.a.aivencloud.com"
port = 23413
username = 'avnadmin'
password = "AVNS_o9Uf-iAsqNySoqOtmNf"
database = 'db_mysql'

# koneksi ke database MySQL
conn = mysql.connector.connect(
    host=hostname,
    port=port,
    user=username,
    password=password,
    database=database
)

# Membaca data dari database menggunakan Pandas
query = "SELECT SepalLengthCm, SepalWidthCm FROM dataoutlier"
cursor = conn.cursor()
cursor.execute(query)
rows = cursor.fetchall()
x = [[row[0], row[1]] for row in rows]

# Menutup koneksi ke database
conn.close()

# Membuat model LOF
lof = LocalOutlierFactor(n_neighbors=13, contamination=0.03)

# Melakukan prediksi outlier
y_pred = lof.fit_predict(x)

# Mendapatkan LOF scores
lof_scores = lof.negative_outlier_factor_

# Menampilkan data dengan outlier yang diidentifikasi
outlier_indices = [i for i, label in enumerate(y_pred) if label == -1]
outlier_data = [(x[i], lof_scores[i]) for i in outlier_indices]

# Sorting outlier_data by LOF scores
outlier_data.sort(key=lambda x: x[1])

# Menampilkan outlier dengan LOF scores dan rank
for rank, (point, score) in enumerate(outlier_data, start=1):
    print(f"Rank {rank}: LOF Score: {score}, Data: {point}")

# Visualisasi data dengan outlier yang diidentifikasi
outlier_points = [point for point, _ in outlier_data]

plt.scatter([point[0] for point in x], [point[1] for point in x])
plt.scatter([point[0] for point in outlier_points], [point[1] for point in outlier_points], color='r')
plt.show()
```
